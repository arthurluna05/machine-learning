```python
import pandas as pd

uri = "https://raw.githubusercontent.com/alura-cursos/introducao-a-data-science/refs/heads/master/aula2.1/movies.csv"
filmes = pd.read_csv(uri) # Lê os dados do CSV e transforma em um DataFrame
filmes.columns = ["filmeId", "titulo", "generos"] # Renomeia as colunas do DataFrame
filmes = filmes.set_index('filmeId') # Define a coluna 'filmeId' como índice do DataFrame, que é substituído por números sequenciais
filmes = filmes.join(filmes['generos'].str.get_dummies()).drop('generos', axis=1) # Cria colunas para cada gênero e remove a coluna 'generos'
filmes['ano'] = filmes['titulo'].str.extract(r'.*\((\d+)\)') # Extrai o ano do título do filme e cria uma nova coluna 'ano'
flmes = filmes.dropna() # Remove linhas com valores ausentes
filmes.head()       
```

```python
uri = "https://raw.githubusercontent.com/alura-cursos/introducao-a-data-science/refs/heads/master/aula1.2/ratings.csv"
notas = pd.read_csv(uri) # Lê as notas do CSV e transforma em um DataFrame
notas.columns = ["usuarioId", "filmeId", "nota", "momento"] # Renomeia as colunas do DataFrame
arredondadas = notas['nota'].round(1) # Arredonda as notas para o inteiro mais próximo
notas.head()
```

```python
medias = notas.groupby("filmeId")["nota"].mean() # Calcula a média das notas para cada filme
filmes = filmes.join(medias).dropna().sort_values("nota", ascending=False).rename(columns={"nota": "media"}) # Adiciona a média ao DataFrame filmes, remove linhas com valores ausentes, ordena por média decrescente e renomeia a coluna

total = notas.groupby('filmeId')["momento"].count() # Conta o número de avaliações para cada filme
filmes = filmes.join(total) # Adiciona a contagem ao DataFrame filmes
filmes = filmes.rename(columns={"momento": "total_de_avaliacoes"}) # Renomeia a coluna 'momento' para 'total_de_avaliacoes'
filmes = filmes.join([medias, total]) # Filtra filmes com pelo menos 50 avaliações
filmes['media_categoria'] = (filmes['media']).round(1).values # Arredonda a média para o inteiro mais próximo e cria uma nova coluna 'media_categoria'

random_filmes = filmes.sample(10) # Seleciona 10 filmes aleatórios do DataFrame
random_filmes.head()
```
